# Reading Data


```python
import pandas as pd
data = pd.read_csv("dataset.csv")
data.head()
```

# Splitting Data

```python
len(data)
```

```python
train_data = data[:22000]
test_data = data[22000:]
```

```python
print(len(train_data))
print(len(test_data))
```

# Pre-Processing 

```python
del train_data['id'] # I droped the id column because it's relevant to the problem and it should not effect on training
del test_data['id']
train_data.head()
```

```python
for col in train_data.columns:
  if train_data[col].isnull().any() == True:
    print(col)
```

```python
for col in test_data.columns:
  if test_data[col].isnull().any() == True:
    print(col)
```

```python
#filling null values
train_data['TargetProId'].fillna('0', inplace=True)
train_data['Price'].fillna(method='ffill', inplace=True)

test_data['TargetProId'].fillna('0', inplace=True)
test_data['Price'].fillna(method='ffill', inplace=True)

train_data.isnull().any()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
for column_name in train_data.columns:
    if train_data[column_name].dtype == object:
        train_data[column_name] = le.fit_transform(train_data[column_name])
    else:
        pass
train_data
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
for column_name in test_data.columns:
    if test_data[column_name].dtype == object:
        test_data[column_name] = le.fit_transform(test_data[column_name])
    else:
        pass
test_data
```

```python
import numpy
X = train_data
Y = train_data['isFraud']

X.drop( columns='isFraud', axis=1, inplace=True)

X.to_numpy()
Y.to_numpy()
print(X)
```

# Training Model
Supervised Classification Models

```python
#SVM
from sklearn import svm

clf = svm.SVC()
clf.fit(X, Y)
```

```python
X_test = test_data
Y_test = test_data['isFraud']

X_test.drop( columns='isFraud', axis=1, inplace=True)

X_test.to_numpy()
Y_test.to_numpy()

y_pred = clf.predict(X_test)
```

# Evaluation

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test, y_pred)
```
